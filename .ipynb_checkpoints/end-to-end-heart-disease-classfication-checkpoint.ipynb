```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X, y
```

```python
np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

I am going to import 3 classifiers to try which one is the best.
1. Logistic Regression
2. RandomForestClassifier
3. KNeigborsClassifier

```python
models = {"Logistic Reg": LogisticRegression(),
         "KNN": KNeighborsClassifier(),
         "Random Forest": RandomForestClassifier()}

def score_models(models, X_train, X_test, y_train, y_test):
    np.random.seed(42)
    model_scores = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = score_models(models, X_train, X_test, y_train, y_test)
model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=["accuracy"])
model_compare.T.plot.bar();
```

### HYPERPARAMETER TUNING

```python
train_scores = []
test_scores = []

neighbors = range(1, 21)
knn = KNeighborsClassifier()

for i in neighbors:
    knn.set_params(n_neighbors=i)
    knn.fit(X_train, y_train)
    train_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))
```

```python
train_scores
```

```python
test_scores
```

```python
plt.plot(neighbors, train_scores, label="Train Score")
plt.plot(neighbors, test_scores, label="Test Score")
plt.xticks(np.arange(1,21,1))
plt.xlabel("Number of Neighbors")
plt.ylabel("Model Score")
plt.legend();
print(f"Max Score of tests is: {max(test_scores)*100:.2f}%")
```

In the below cell i tried to implement some different parameters to improve my model but it seems like it didn't work. Thus, i will try this with GridSearchCV

```python
log_reg_grid = {"C": [100, 10, 1.0, 0.1, 0.01],
               "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"],
                "penalty": [None , "l1", "l2", "elasticnet"]
               }
rf_grid = {"n_estimators": np.arange(10,1000,50),
          "max_depth":[None, 3, 5, 10],
          "min_samples_split": np.arange(2, 20,2),
          "min_samples_leaf":np.arange(1,20,2),
          }
```

```python
np.random.seed(42)

rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                               param_distributions=log_reg_grid,
                               cv=5,
                               n_iter=20,
                               verbose=True)

rs_log_reg.fit(X_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(X_test, y_test)
```

LogisticRegression results are same i will try to tune different params to improve model.

```python
np.random.seed(42)

rs_rf = RandomizedSearchCV(RandomForestClassifier(),
                          param_distributions=rf_grid,
                          cv=5,
                          n_iter=20,
                          verbose=True)
rs_rf.fit(X_train, y_train)
```

```python
rs_rf.best_params_
```

```python
rs_rf.score(X_test, y_test)
```

So, even i tuned rfc very well it's still behind logistic regression so i decided to choose logistic regression to make my main model.

```python
log_reg_grid = {"C": np.logspace(-5,5,20),
               "solver": ["newton-cg", "liblinear", "sag", "saga"],
               "penalty": [None, "l1", "l2", "elasticnet"]}
gs_log_reg = GridSearchCV(LogisticRegression(),
                         param_grid=log_reg_grid,
                         cv=5,
                         verbose=True)
gs_log_reg.fit(X_train, y_train);
```

```python
sns.set(font_scale=1.5)
def plot_conf_mat(y_test, y_preds):
    """
    PLOTS CONFUSION MATRIX
    """
    fig, ax = plt.subplots(figsize=(5,5
                                   ))
    ax= sns.heatmap(confusion_matrix(y_test, y_preds),
                   annot=True,
                   cbar=True,
                    cmap="rocket_r"
                   )
    
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
```

```python
plot_conf_mat(y_test, y_preds)
```

```python
print(classification_report(y_test, y_preds))
```

```python
clf = LogisticRegression(C=0.1623776739188721, penalty="l2", solver='liblinear')
```

### Cross validated accuracy, precision and recall

```python
np.random.seed(42)
cv_acc = cross_val_score(clf,
                        X,
                        y,
                        cv=5,
                        scoring="accuracy")
cv_acc = cv_acc.mean()
cv_acc
```

```python
np.random.seed(42)
cv_pre = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring ="precision")

cv_pre = cv_pre.mean()
```

```python
np.random.seed(42)
cv_rec = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring ="recall",
                         n_jobs=1)
cv_rec = cv_rec.mean()
```

```python
np.random.seed(42)
cv_f1 = cross_val_score(clf,
                         X,
                         y,
                         cv=5,
                         scoring ="f1")

cv_f1 = cv_f1.mean()
```

```python
cv_metrics = pd.DataFrame({"Accuracy": cv_acc,
                           "Precision": cv_pre,
                           "Recall": cv_rec,
                           "F1": cv_f1},
                           index=[0])

cv_metrics.T.plot.bar(title="Cross-validated classification metrics",
                     legend=False);
```
